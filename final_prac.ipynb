```python
import pyspark

spark = pyspark.sql.SparkSession\
    .builder\
    .master('local')\
    .appName('prac')\
    .config(conf=pyspark.SparkConf())\
    .getOrCreate()
```

```python
_rdd = spark.sparkContext.textFile('./week10text.txt')
_rdd = _rdd.filter(lambda x: x != '').map(lambda x: [x])

df = spark.createDataFrame(_rdd, ['sent'])
df.show()
```

### 단어로 분리

```python
from pyspark.ml.feature import Tokenizer

tokenizer = Tokenizer(inputCol='sent', outputCol='words')
tokDf = tokenizer.transform(df)
tokDf.show()
```

### 불용어 구성, 한 글자로 된 단어 찾아내 구성

```python
stopwords = [
    u"또한",
    u"위해",
    u"참",
    u"더",
    u"그러나",
    u"그",
    u"특히"
]

from pyspark.ml.feature import StopWordsRemover

stop = StopWordsRemover(inputCol='words', outputCol='nostopwords')
stop.setStopWords(stopwords)

stopDf = stop.transform(tokDf)
stopDf.show()
```

### TF-IDF 계산 및 출력

```python
from pyspark.ml.feature import HashingTF, IDF

hashTF = HashingTF(inputCol='nostopwords', outputCol='hash')
hashDf = hashTF.transform(stopDf)

hashDf.show()
```

```python
idf = IDF(inputCol='hash', outputCol='idf')
idfModel = idf.fit(hashDf)
idfDf = idfModel.transform(hashDf)

idfDf.select(['nostopwords', 'hash', 'idf']).show()
```

### TF-IDF column을 feature로 구성, 출력

```python
from pyspark.mllib.regression import LabeledPoint
import pyspark.sql.functions as F

idfDf = idfDf.withColumnRenamed('idf', 'features')
idfDf.select('features').rdd.collect()[0]
```

## K-Means Clustering

```python
from sklearn.datasets import make_blobs
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=1000, centers=4, cluster_std=0.8, random_state=1)

uX0, uX1 = np.mean(X, axis=0)
sX0, sX1 = np.std(X, axis=0)
```

```python
colors = ['blue', 'red', 'yellow', 'green']

plt.scatter(X[y==0, 0], X[y==0, 1], color=colors[0])
plt.scatter(X[y==1, 0], X[y==1, 1], color=colors[1])
plt.scatter(X[y==2, 0], X[y==2, 1], color=colors[2])
plt.scatter(X[y==3, 0], X[y==3, 1], color=colors[3])
plt.show()
```

```python
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans

va = VectorAssembler(inputCols=['X0', 'X1'], outputCol='features')
kmeans = KMeans(k=4)
```

```python
_rdd = spark.sparkContext.parallelize(X)
_rdd = _rdd.map(lambda x: [float(x_) for x_ in x])
_rdd.collect()

_df = _rdd.toDF(["X0", "X1"])
df = va.transform(_df)

kmModel = kmeans.fit(df)
```

```python
kmDf = kmModel.transform(df)
kmDf.show()
```

```python
kmPandas = kmDf.toPandas()

for i in range(4):
    plt.scatter(kmPandas[kmPandas["prediction"]==i]["X0"], 
                kmPandas[kmPandas["prediction"]==i]["X1"], 
                color=colors[i])
plt.show()
```

# Regression Math

```python
import numpy as np

x = np.array([1, 2, 3, 4])
y = np.array([6, 5, 7, 10])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(x, y)
```

```python
import sympy as sp
```

```python
b0, b1 = sp.symbols('b0 b1')

f = 0 
for i in range(len(y)):
    f += (y[i] - (b0 + b1 * x[i]))**2

f
```

```python
sp.diff(f, b0)
```

```python
sp.diff(f, b1)
```

```python
res = sp.solve([sp.diff(f, b0), sp.diff(f, b1)], [b0, b1])
```

```python
b = (res[b0], res[b1])
```

```python
b
```

```python
yhat = b[1] * x + b[0]
```

```python
yhat
```

```python
y - yhat
```

```python
RSS = sum((y-yhat)**2)
TSS = sum((y-np.mean(y))**2)
R_squared = 1 - RSS/TSS
R_squared
```

```python
yhat = w0 + w1*x
```

```python
RSS = sum((y-yhat)**2)
TSS = sum((y-np.mean(y))**2)
R_squared = 1 - RSS/TSS
```

```python
abaloneDf = spark.createDataFrame(abalone)
```

```python
abaloneDf.printSchema()
```

```python
abaloneDf.groupBy('sex').agg({"rings":"Avg"}).show()
```

```python
(trainDf, testDf) = abaloneDf.randomSplit([0.6, 0.4])
```

```python
trainDf.count()
```

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler

si = StringIndexer(inputCol='sex', outputCol='sexIndex')
ohe = OneHotEncoder(inputCol='sexIndex', outputCol='sexVec')
va = VectorAssembler(inputCols=['sexVec', 'length', 'diameter', 'height', 'whole weight',
                                'shucked weight', 'viscera weight', 'shell weight'
                               ], outputCol='features')
```

```python
from pyspark.ml.regression import LinearRegression

lr = LinearRegression(featuresCol='features', labelCol='rings', maxIter=10, regParam=0.3, elasticNetParam=0.8)
```

```python
from pyspark.ml import Pipeline

pipeline = Pipeline(stages=[si, ohe, va, lr])
model = pipeline.fit(trainDf)
modelTrainDf = model.transform(trainDf)
```

```python
modelTrainDf.groupBy('rings').agg({'prediction':'avg'}).show()
```

```python
from pyspark.ml.evaluation import RegressionEvaluator

ev = RegressionEvaluator(labelCol='rings', predictionCol='prediction', metricName='rmse')
rmse = ev.evaluate(modelTrainDf)
```
