## 문제 2: RDD를 사용해서 단어빈도 계산

다음은 일부 수정된 노래가사이다.

````python
yesterday=[
    ["Yesterday 어제는"],
    ["All my troubles seemed to far away 내 모든 문제들이 참 멀리 있는 것 같았는데"],
    ["Now it looks as though they're here to stay 이제는 그 모든 문제들이 여기에 있는 것처럼 보여"],
    ["Oh, I believe in yesterday 오, 어제가 좋았는데"],
    ["Suddenly 갑자기 그-림-자-가 2021 그-림-자-가 2021"],
    ["I'm not half the man I used to be 난 예전 내 모습의 반도 못한 사람이 되어버렸어"],
    ["There's a shadow hanging over me 내 위에 그림자가 드리워져 있네"],
    ["Oh!, yesterday came suddenly 오!, 이제는 너무 갑자기 와버렸어.."],
]
````

1) 단어빈도를 계산해서 출력

출력 예시: [('Yesterday', 1),('어제는', 1),('All', 1),('my', 1),...]

````python
import pyspark

spark = pyspark.sql.SparkSession\
    .builder\
    .master("local")\
    .appName("midterm")\
    .config(conf=pyspark.SparkConf())\
    .getOrCreate()
````

````python
wcRdd = spark.sparkContext.parallelize(yesterday)
_wcRdd = wcRdd\
    .flatMap(lambda x: x[0].split())\
    .map(lambda x: (x, 1))\
    .reduceByKey(lambda x, y: x + y)

print(_wcRdd.collect())
````

2) 빈도순으로 출력, 빈도 1이하는 제외.

````python
__wcRdd = wcRdd\
    .flatMap(lambda x: x[0].split())\
    .map(lambda x: (x, 1))\
    .reduceByKey(lambda x, y: x + y)\
    .filter(lambda x: x[1] > 1)\
    .map(lambda x: (x[1], x[0]))\
    .sortByKey(False)\
    .map(lambda x: (x[1], x[0]))

print(__wcRdd.collect())
````

3) 소문자, 하이픈, 느낌표, 숫자를 제외하고 단어빈도 계산해서 출력 (빈도 1이하는 제외)

* 소문자: 'Yesterday', 'yesterday'를 동일한 단어로 계산

* 하이픈: '그림자가', '그-림-자-가'를 동일한 단어로 계산

* 느낌표: 'oh', 'oh!'를 동일한 단어로 계산

* 숫자: '2021'는 제외하고 계산

````python
import re

regex1 = re.compile(r"[-!,]")
___wcRdd = wcRdd\
    .flatMap(lambda x: x[0].split())\
    .map(lambda x: regex1.sub("", x))\
    .map(lambda x: x.lower())\
    .filter(lambda x: not str(x).isdecimal())\
    .map(lambda x: (x, 1))\
    .reduceByKey(lambda x, y: x + y)\
    .filter(lambda x: x[1] > 1)\
    .map(lambda x: (x[1], x[0]))\
    .sortByKey(False)\
    .map(lambda x: (x[1], x[0]))

print(___wcRdd.collect())
````

4) 1글자 단어 (예를 들어 'i', '내') 제외하고 단어빈도 계산 (빈도 1이하는 제외)

````python
import re

regex1 = re.compile(r"[-!,]")
____wcRdd = wcRdd\
    .flatMap(lambda x: x[0].split())\
    .map(lambda x: regex1.sub("", x))\
    .map(lambda x: x.lower())\
    .filter(lambda x: not str(x).isdecimal())\
    .filter(lambda x: len(x) > 1)\
    .map(lambda x: (x, 1))\
    .reduceByKey(lambda x, y: x + y)\
    .filter(lambda x: x[1] > 1)\
    .map(lambda x: (x[1], x[0]))\
    .sortByKey(False)\
    .map(lambda x: (x[1], x[0]))

print(____wcRdd.collect())
````

1) 전체 건수 출력. null 값을 제외한 null 건수 출력 (null이 남아있는지 확인하기 위해)

````python
!ls data/*.json
````

````python
covidDf = spark.read.json("data/서울특별시-코로나19-자치구별-확진자-발생동향.json")

import pandas as pd

covidPd = covidDf.toPandas()
covidPd
````

값이 이상합니다. json으로 따로 읽어서 분석을 하고 DataFrame으로 다시 만듭니다.

````python
covidData = ""
with open("data/서울특별시-코로나19-자치구별-확진자-발생동향.json", "r") as f:
    for line in f.readlines():
        covidData += line
    covidJson = json.loads(covidData)

covidJson.keys()
````

````python
covidJson["DESCRIPTION"]
````

````python
covidJson["DATA"][:10]
````

"DATA" 에 원하는 정보가 있으므로 이를 사용합니다.

위 정보들을 하나하나 Row로 변형하여 DataFrame으로 만듭니다.

````python
from pyspark.sql import Row
import pyspark.sql.functions as F

covidRows = [Row(**x) for x in covidJson["DATA"]]

_covidDf = spark.createDataFrame(covidRows)
_covidDf.toPandas()
````

null 값을 제거하기전의 count는 아래와 같습니다.

````python
_covidDf.select([F.count(F.col(c)).alias(c) for c in _covidDf.columns]).toPandas()
````

 예쁘게 출력되는 것을 알 수 있습니다. 
 
 아래 None으로 null 값들이 존재함을 알 수 있고 해당 record들을 삭제해줍니다.

````python
covidDfNullDrop = _covidDf.na.drop()
covidDfNullDrop.toPandas()
````
