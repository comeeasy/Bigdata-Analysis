# 문제 7: 버스 승하차 인원 월별 정류장별 인원

```python
import os
from src import myOauth as OAuth

public_keys = OAuth.getKey('data/key.properties')
key = str(public_keys['dataseoul'])
```

```python
import requests
import re

bus_stop_base_url = "http://openAPI.seoul.go.kr:8088/"+key+"/xml/CardBusTimeNew/1/5/202107/7016"
data = requests.get(bus_stop_base_url).text
```

```python
print(data)
```

### Regular Expression으로 XML parsing

```python
p = re.compile('<BUS_STA_NM>(.+?)</BUS_STA_NM>')
res = p.findall(data)
for item in res:
    print(item)
```

```python
p = re.compile("<.*_NUM>(\d+)</.*_NUM>")
res = p.findall(data)
for item in res:
    print(item, end=", ")
```
