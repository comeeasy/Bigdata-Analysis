# 인증키 myOAuth

```python
import src.myOauth as OAuth

public_keys = OAuth.getKey("data/key.properties")
print(public_keys)
```

```python
base_url = "http://openAPI.seoul.go.kr"
port_number = "8088"

access_key = public_keys["dataseoul"]
data_type = "json"
service = "SearchSTNBySubwayLineService"
start_idx = "1"
end_idx = "10"
station_code = ""
station_name = ""
line_name = "2호선"
params = "/".join([access_key, data_type, service, start_idx, end_idx, 
                   station_code, station_name, line_name])

url = base_url+":"+port_number+"/"+params
print(url)
```

```python
# get total data

import requests

r = requests.get(url)
data = r.json()

print(data)
print(data["SearchSTNBySubwayLineService"]["list_total_count"])

end_idx = str(data["SearchSTNBySubwayLineService"]["list_total_count"])
```

```python
# get whole data
params = "/".join([access_key, data_type, service, start_idx, end_idx,
                   station_code, station_name, line_name])

url = base_url+":"+port_number+"/"+params
r = requests.get(url)
data = r.json()

total_list_num = len(data["SearchSTNBySubwayLineService"]["row"])
print(f"total # of data: {total_list_num}")
for idx, info in enumerate(data["SearchSTNBySubwayLineService"]["row"]):
    print(f"idx: {idx+1:3d} : {info}")
```

# XML 형식으로 받아오기

```python
data_type = 'xml'

params = "/".join([access_key, data_type, service, start_idx, end_idx, 
                   station_code, station_name, line_name])

url = base_url+":"+port_number+"/"+params
print(url)
```

```python
import lxml
import lxml.etree
import requests

data = requests.get(url).text
tree = lxml.etree.fromstring(data.encode("utf-8"))
print(tree)
```

### XPath 이용하기 

'//STATION_NM' 은 어느 계층에 있던지 간에 일치하는 태그를 찾는다.

```python
for node in tree.xpath("//STATION_NM"):
    print(node.text)
```

```python
for node in tree.xpath("//list_total_count"):
    print(node.text)
```
