## 변수의 성격

##### 변수는 크게 **숫자**, **문자** 로 구분된다.

### Categorical (=nominal) Variable
    - 순서가 없이 이름이 있는 변수를 말한다. 
    - Ex. [검은색, 분홍색, 흰색], [남자, 여자]
### Ordinal Variable
    - 순서가 있는 이름을 말한다.
    - Ex. [초등학교, 중학교, 고등학교], [차상위, 중위, 상위]
### Interval Variable
    - 간격이 있는 변수
    - Ex. ['Bad', 'soso' ,'good']
### Continuous Variavle
    - 연속적인 변수
    - Ex. 키, 몸무게

## Sampling

```python
import random

random.seed(3)
```

```python
import numpy as np
np.random.seed(3)

np.random.random((3, 2))
```

- `np.random.rand`는 uniform distrubution<br>
- `np.random.randn`은 standard normal distribution

```python
random.sample([1, 2, 3, 4, 5, 6], 2)
```

### Uniform Distribution

```python
dice = [random.randrange(1, 6+1) for _ in range(10000)]
print(dice)
```

#### Histogram

```python
for i in range(1, 7):
    print(f"Dice {i}: {dice.count(i)}회")
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.hist(dice)
plt.show()
```

```python
plt.hist(dice, bins=np.arange(0.5, 7))
plt.show()
```

```python
import pandas as pd

pd.Series(dice).hist(bins=np.arange(0.5, 7))
```

### Describe

통계량을 구해보자. describe()는 기술통계량 평균, 최소값, 표준편차, 건수를 출력한다.<br> 
DataFrame 또는 컬럼을 Series로 변환한 후 describe() 를 적용할 수 있다.

```python
import pandas as pd

s = pd.Series(dice)
s.describe()
```

### 정규 분포

#### Sampling of Gaussian distribution

```python
import numpy as np

comp1 = np.random.normal(0, 1, size=200000)
comp2 = np.random.normal(10, 2, size=200000)
```

```python
x = np.random.randn(100)
count, binends = np.histogram(x)
print(count, binends)
```

```python
plt.hist(x)
```

```python
pd.Series(np.concatenate([comp1, comp2])).plot()
```

```python
pd.Series(np.concatenate([comp1, comp2])).hist()
```
